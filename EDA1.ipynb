```python
import xml.etree.ElementTree as et
tree=et.parse("C:\\Users\\dhanashree.mokashi\\Desktop\\EDA Datasets\\EDA_Master\\Data\\reviews.xml")
```

```python
tree
```

```python
tree.findall
```

```python
root=tree.getroot()
```

```python
root
```

```python
root.attrib
```

```python
restDict={}
reviewdict={}
specialFeatureList=[]
reviewList=[]
```

```python
for i in root:
    restDict.update({i.tag:i.text})
    for j in i:
        if(i.tag=="SpecialFeatures"):
            specialFeatureList.append(j.text)
            restDict.update({"SpecialFeatures":specialFeatureList})
        if(i.tag=="Reviews"):
            reviewdict={}
            reviewdict.update({"ReviewId": j.attrib["id"]})
            for r in j:
                reviewdict.update({r.tag:r.text})
            reviewList.append(reviewdict)
            restDict.update({"Review List":reviewList})
restDict
```

```python
keys=restDict.keys()
for k in keys:
    print(restDict[k])
```
